```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
from sklearn.decomposition import PCA, KernelPCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
from sklearn.datasets import make_moons, make_circles



def load_wine_df():
    cols = ['Class label', 'Alcohol', 'Malic acid', 'Ash', 'Alcalinity of ash',
            'Magnesium', 'Total phenols', 'Flavanoids', 'Nonflavanoid phenols',
            'Proanthocyanins', 'Color intensity', 'Hue',
            'OD280/OD315 of diluted wines', 'Proline']
    try:
        df = pd.read_csv(
            'https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data',
            header=None)
        df.columns = cols
        return df
    except Exception:
        from sklearn.datasets import load_wine
        data = load_wine()
        df = pd.DataFrame(data=np.c_[data.target + 1, data.data],
                          columns=['Class label'] + data.feature_names)

        rename = {
            'alcohol':'Alcohol','malic_acid':'Malic acid','ash':'Ash',
            'alcalinity_of_ash':'Alcalinity of ash','magnesium':'Magnesium',
            'total_phenols':'Total phenols','flavanoids':'Flavanoids',
            'nonflavanoid_phenols':'Nonflavanoid phenols','proanthocyanins':'Proanthocyanins',
            'color_intensity':'Color intensity','hue':'Hue',
            'od280/od315_of_diluted_wines':'OD280/OD315 of diluted wines',
            'proline':'Proline'
        }
        return df.rename(columns=rename)

from matplotlib.colors import ListedColormap
def plot_decision_regions(X, y, classifier, resolution=0.02, title=''):
    markers = ('s', 'x', 'o', '^', 'v')
    colors = ('red', 'blue', 'lightgreen', 'gray', 'cyan')
    cmap = ListedColormap(colors[:len(np.unique(y))])

    x1_min, x1_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    x2_min, x2_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx1, xx2 = np.meshgrid(np.arange(x1_min, x1_max, resolution),
                           np.arange(x2_min, x2_max, resolution))
    Z = classifier.predict(np.c_[xx1.ravel(), xx2.ravel()])
    Z = Z.reshape(xx1.shape)

    plt.figure()
    plt.contourf(xx1, xx2, Z, alpha=0.4, cmap=cmap)
    plt.xlim(xx1.min(), xx1.max())
    plt.ylim(xx2.min(), xx2.max())
    for idx, cl in enumerate(np.unique(y)):
        plt.scatter(X[y == cl, 0], X[y == cl, 1],
                    alpha=0.6, c=[cmap(idx)], edgecolor='black',
                    marker=markers[idx], label=str(cl))
    plt.legend(loc='best')
    if title:
        plt.title(title)
    plt.xlabel('Component 1')
    plt.ylabel('Component 2')
    plt.show()


df_wine = load_wine_df()
X, y = df_wine.iloc[:, 1:].values, df_wine.iloc[:, 0].values

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, stratify=y, random_state=0)

sc = StandardScaler()
X_train_std = sc.fit_transform(X_train)
X_test_std = sc.transform(X_test)



cov_mat = np.cov(X_train_std.T)
eigen_vals, eigen_vecs = np.linalg.eigh(cov_mat)

sorted_idx = np.argsort(eigen_vals)[::-1]
eigen_vals = eigen_vals[sorted_idx]
eigen_vecs = eigen_vecs[:, sorted_idx]


W_pca2 = eigen_vecs[:, :2]
X_train_pca_scratch = X_train_std @ W_pca2
X_test_pca_scratch = X_test_std @ W_pca2

tot = eigen_vals.sum()
var_exp = eigen_vals / tot
cum_var_exp = np.cumsum(var_exp)

plt.figure()
plt.bar(range(1, len(var_exp)+1), var_exp, alpha=0.5, align='center', label='Individual explained variance')
plt.step(range(1, len(cum_var_exp)+1), cum_var_exp, where='mid', label='Cumulative explained variance')
plt.ylabel('Explained variance ratio')
plt.xlabel('Principal components')
plt.legend(loc='best')
plt.tight_layout()
plt.show()


n95 = int(np.searchsorted(cum_var_exp, 0.95) + 1)
print(f"[PCA Scratch] Components to reach 95% variance: {n95}")

pca = PCA(n_components=2, random_state=0)
X_train_pca = pca.fit_transform(X_train_std)
X_test_pca = pca.transform(X_test_std)
print(f"[PCA Sklearn] Explained variance ratios: {pca.explained_variance_ratio_}, cumulative={pca.explained_variance_ratio_.sum():.3f}")


lr_pca = LogisticRegression(multi_class='ovr', random_state=1, solver='lbfgs', max_iter=200)
lr_pca.fit(X_train_pca, y_train)
plot_decision_regions(X_train_pca, y_train, classifier=lr_pca, title='Logistic Regression on PCA (train)')
print(f"[PCA] Test Accuracy (LR): {lr_pca.score(X_test_pca, y_test):.3f}")


# Part 2: LDA (Scratch + scikit-learn)



labels = np.unique(y_train)
d = X_train_std.shape[1]


S_W = np.zeros((d, d))
mean_vecs = []
for cl in labels:
    Xc = X_train_std[y_train == cl]
    mean_vecs.append(Xc.mean(axis=0))
    S_W += np.cov(Xc.T, bias=False)


mean_overall = X_train_std.mean(axis=0).reshape(d, 1)
S_B = np.zeros((d, d))
for cl, mv in zip(labels, mean_vecs):
    n = X_train_std[y_train == cl].shape[0]
    mv = mv.reshape(d, 1)
    S_B += n * (mv - mean_overall) @ (mv - mean_overall).T



eps = 1e-6
eig_vals_lda, eig_vecs_lda = np.linalg.eig(np.linalg.pinv(S_W + eps*np.eye(d)) @ S_B)


idx = np.argsort(np.real(eig_vals_lda))[::-1]
eig_vals_lda = np.real(eig_vals_lda[idx])
eig_vecs_lda = np.real(eig_vecs_lda[:, idx])


W_lda2 = eig_vecs_lda[:, :2]
X_train_lda_scratch = X_train_std @ W_lda2
X_test_lda_scratch  = X_test_std  @ W_lda2

lda = LDA(n_components=2)
X_train_lda = lda.fit_transform(X_train_std, y_train)
X_test_lda = lda.transform(X_test_std)


lr_lda = LogisticRegression(multi_class='ovr', random_state=1, solver='lbfgs', max_iter=200)
lr_lda.fit(X_train_lda, y_train)
plot_decision_regions(X_train_lda, y_train, classifier=lr_lda, title='Logistic Regression on LDA (train)')
print(f"[LDA] Test Accuracy (LR): {lr_lda.score(X_test_lda, y_test):.3f}")



lr_orig = LogisticRegression(multi_class='ovr', random_state=1, solver='lbfgs', max_iter=500)
lr_orig.fit(X_train_std, y_train)
print(f"[Original Std] Test Accuracy (LR): {lr_orig.score(X_test_std, y_test):.3f}")


svm_orig = SVC(kernel='linear', C=1.0, random_state=0)
svm_orig.fit(X_train_std, y_train)
svm_pca = SVC(kernel='linear', C=1.0, random_state=0).fit(X_train_pca, y_train)
svm_lda = SVC(kernel='linear', C=1.0, random_state=0).fit(X_train_lda, y_train)
print(f"[Original Std] Test Accuracy (SVM-linear): {svm_orig.score(X_test_std, y_test):.3f}")
print(f"[PCA 2D]      Test Accuracy (SVM-linear): {svm_pca.score(X_test_pca, y_test):.3f}")
print(f"[LDA 2D]      Test Accuracy (SVM-linear): {svm_lda.score(X_test_lda, y_test):.3f}")


# Part 3: Kernel PCA (Scratch + scikit-learning


from scipy.spatial.distance import pdist, squareform

def rbf_kernel_pca(X, gamma=15.0, n_components=2):
    """
    RBF Kernel PCA:
    1) Compute pairwise squared Euclidean distances
    2) Build RBF kernel matrix K
    3) Center K
    4) Eigen-decompose centered K
    5) Return top n_components projections
    """

    sq_dists = pdist(X, 'sqeuclidean')

    K = np.exp(-gamma * squareform(sq_dists))

    N = K.shape[0]
    one_n = np.ones((N, N)) / N
    K_centered = K - one_n @ K - K @ one_n + one_n @ K @ one_n

    eigvals, eigvecs = np.linalg.eigh(K_centered)

    X_pc = np.column_stack([eigvecs[:, -i] for i in range(1, n_components+1)])
    return X_pc


Xm, ym = make_moons(n_samples=300, noise=0.05, random_state=123)
Xm_kpca = rbf_kernel_pca(Xm, gamma=15, n_components=2)

plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.scatter(Xm[ym==0,0], Xm[ym==0,1], alpha=0.5, marker='^')
plt.scatter(Xm[ym==1,0], Xm[ym==1,1], alpha=0.5, marker='o')
plt.title('Original Half-Moon Data')
plt.subplot(1,2,2)
plt.scatter(Xm_kpca[ym==0,0], Xm_kpca[ym==0,1], alpha=0.5, marker='^')
plt.scatter(Xm_kpca[ym==1,0], Xm_kpca[ym==1,1], alpha=0.5, marker='o')
plt.title('RBF Kernel PCA (γ=15)')
plt.tight_layout()
plt.show()


Xc, yc = make_circles(n_samples=600, random_state=123, noise=0.05, factor=0.3)
Xc_kpca = rbf_kernel_pca(Xc, gamma=15, n_components=2)

plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.scatter(Xc[yc==0,0], Xc[yc==0,1], alpha=0.5, marker='^')
plt.scatter(Xc[yc==1,0], Xc[yc==1,1], alpha=0.5, marker='o')
plt.title('Original Circles Data')
plt.subplot(1,2,2)
plt.scatter(Xc_kpca[yc==0,0], Xc_kpca[yc==0,1], alpha=0.5, marker='^')
plt.scatter(Xc_kpca[yc==1,0], Xc_kpca[yc==1,1], alpha=0.5, marker='o')
plt.title('RBF Kernel PCA (γ=15)')
plt.tight_layout()
plt.show()

# KPCA with scikit-learn (half-moons for comparison)
kpca = KernelPCA(n_components=2, kernel='rbf', gamma=15, fit_inverse_transform=False, random_state=0)
Xm_kpca_sk = kpca.fit_transform(Xm)

plt.figure()
plt.scatter(Xm_kpca_sk[ym==0,0], Xm_kpca_sk[ym==0,1], alpha=0.5, marker='^')
plt.scatter(Xm_kpca_sk[ym==1,0], Xm_kpca_sk[ym==1,1], alpha=0.5, marker='o')
plt.title('scikit-learn KPCA (γ=15) — Half-Moons')
plt.show()



def demo_gamma_values(X, y, gammas=(0.01, 1, 15, 100)):
    for g in gammas:
        Xk = rbf_kernel_pca(X, gamma=g, n_components=2)
        plt.figure()
        plt.scatter(Xk[y==0,0], Xk[y==0,1], alpha=0.5, marker='^')
        plt.scatter(Xk[y==1,0], Xk[y==1,1], alpha=0.5, marker='o')
        plt.title(f'RBF KPCA (γ={g})')
        plt.show()


# demo_gamma_values(Xm, ym, gammas=(0.01, 0.1, 1, 10, 50, 100


print("\n=== Summary ===")
print(f"LR (Original std)  : {accuracy_score(y_test, lr_orig.predict(X_test_std)):.3f}")
print(f"LR (PCA 2D)        : {accuracy_score(y_test, lr_pca.predict(X_test_pca)):.3f}")
print(f"LR (LDA 2D)        : {accuracy_score(y_test, lr_lda.predict(X_test_lda)):.3f}")
print(f"SVM-lin (Original) : {svm_orig.score(X_test_std, y_test):.3f}")
print(f"SVM-lin (PCA 2D)   : {svm_pca.score(X_test_pca, y_test):.3f}")
print(f"SVM-lin (LDA 2D)   : {svm_lda.score(X_test_lda, y_test):.3f}")
print(f"PCA scratch 95% n_components: {n95}")
print("Done.")
```
